```python
#1 - Crie um programa que exiba na tela seu nome e sua idade.
nome = input("Digite seu nome: ")
idade = int(input("Digite a sua idade:"))
print(f' Seu nome é {nome}, sua idade é {idade}.')
```

```python
# 2 - Faça um programa que peça o nome do usuário e exiba uma mensagem de boas-vindas com o nome digitado.
nome = input("Digite seu nome:")
print(f'Bem vindo {nome}.')
```

```python
# 3 - Desenvolva um programa que peça dois números ao usuário e mostre a soma entre eles.
#input() - faz leitura de uma variável em texto
#int(input()) - converte para inteiro
#float(input()) - converte pra decimal

Num1 = int(input("Digite o primeiro numero:"))
Num2 = int(input("Digite o segundo numero:"))

print(f"{Num1} + {Num2} = {Num1 + Num2}")
```

```python
#4 - Crie um programa que peça um número inteiro e mostre seu antecessor e seu sucessor.

Num = int(input("Digite um número:"))

print(f'Antecessor: {Num - 1}, Sucessor {Num + 1}')
```

```python
#5 - Faça um programa que peça o ano de nascimento do usuário e calcule sua idade aproximada.
ano = int(input('Digite o seu ano de nascimento:'))

print(f'Sua idade é de {2026 - ano}.')
```

```python
#6 - Desenvolva um programa que peça duas notas e calcule a média entre elas.
nota1 = int(input('Digite a primeira nota:'))
nota2 = int(input('Digite a segunda nota:'))

print(f'A média entre as suas notas foi de {(nota1 + nota2)/2}')
```

```python
#7 - Crie um programa que peça uma temperatura em graus Celsius e converta para Fahrenheit
tempC = float(input('Digite a têmperatura em Celsius:'))

print(f'A sua têmperatura em fahrenheit é de {(tempC * 1.8)+32}F')
```

```python
#8 - Faça um programa que peça a largura e a altura de um retângulo e calcule sua área.
# :.2f - duas casas após a vírgula, apenas. pode mudar o número.

H = float(input('Digite a  altura do triângulo (cm):'))
L = float(input('Digite a  largura do triângulo (cm):'))

print(f'Á área do seu triângulo é de {(H*L) :.2F} cm2:')
```

```python
#9 - Desenvolva um programa que peça um número inteiro e calcule o seu quadrado
num = int(input('Digite um número:'))

print(f' O quadrado do número escolhido é: {num*num}')
```

```python
#10 - Crie uma calculadora simples que peça dois números e uma operação matemática (+, -, *, /) e mostre o resultado.
#eval() - avalia o texto e executa uma determinada atividade interpretada

num1 = input('DIgite o primeiro número:')
num2 = input('DIgite o segundo número:')
op = input('Digite um operador (+, -, / ou *):')

resultado = eval(num1+op+num2)

print(resultado)
```
