```python
#1 - Peça um número inteiro ao usuário. Se o número for maior que zero, exiba a mensagem "O número é positivo!".
num = int(input('Digite um número: '))
if num > 0:
    print("O número é positivo!")
```

```python
#2 Leia a idade do usuário. Se for maior ou igual a 18, exiba "Maior de idade", caso contrário, exiba "Menor de idade".
idade = int(input('Digite sua idade: '))
if idade >= 18:
    print('Maior de idade')
else:
    print('Menor de idade')
```

```python
#3 Solicite um número inteiro. Verifique se ele é par (divisível por 2) e exiba uma mensagem informando o resultado.
num = int(input('Digite um número: '))
if num%2 == 0:
    print('Número par!')
else:
    print('Número impar!')
```

```python
#4 - Defina uma senha fixa no código (ex: "python123"). Peça a senha ao usuário. Se bater, exiba "Acesso permitido", senão "Acesso negado".
password = input('Digite a senha: ')
if password == 'Python123':
    print('Senha Correta!')
else:
    print('Senha Incorreta!')
```

```python
#5 - Leia a temperatura em graus Celsius. Se for maior que 30, exiba "Está quente hoje!". Caso contrário, exiba "Temperatura agradável."
temperatura = float(input('Digite a temperatura atual (C°):'))
if temperatura > 30:
    print("Está quente hoje!")
else:
    print("Temperatura agradável.")
```

```python
#6 - Defina um saldo fixo (ex: 100.0). Peça ao usuário um valor para saque. Se o valor for menor ou igual ao saldo, exiba "Saque realizado!", caso contrário, "Saldo insuficiente."
saldo = 1000
saque = float(input('Digite o valor de saque: '))
if saque <= saldo:
    print("Saque realizado!")
else:
    print("Saldo insuficiente!")
```

```python
#7 - Leia a nota de um aluno (0 a 10). Se a nota for maior ou igual a 5, exiba "Aprovado". Senão, exiba "Reprovado".
nota = float(input("Digite a sua nota:"))
if nota >= 5:
    print("Aprovado")
else:
    print("Reprovado")
```

```python
#8 - Peça um número ao usuário. Exiba se o número é zero ou diferente de zero usando if/else.
numero = int(input('Digite um número: '))
if numero != 0:
    print('Diferente de 0')
else:
    print('Im Zero!')
```

```python
#9 - Peça um número ao usuário. Sem usar == 0, use o próprio valor como condição no if para exibir "O número é falsy (zero)" ou "O número é truthy (diferente de zero)".
numero = bool(int(input('Digite um número: ')))
if numero is True:
    print('O número é truthy (diferente de zero)')
else:
    print("O número é falsy (zero)")
```

```python
# 10 - Peça ao usuário para digitar a cor do semáforo ("verde" ou outra cor). Se for "verde", exiba "Pode avançar!", caso contrário, "Pare!".
cor = input('Digite a cor do semáforo:')
if cor.upper() == 'VERDE':
    print('Pode avançar!')
else:
    print('Pare!')
```

```python
#11 - Leia o nome do usuário. Se o nome for uma string vazia (falsy), exiba "Nenhum nome digitado.", caso contrário, exiba "Olá, [nome]!" usando f-string.
nome = input('Digite seu nome:')
if bool(nome) is False:
# ou if not nome:
    print("Nenhum nome digitado.")
else:
    print(f'Olá, {nome}!')
```

```python
#12 - Leia dois números inteiros. Exiba qual deles é o maior. Se forem iguais, exiba "Os números são iguais."
intA = int(input('Digite o 1° número: '))
intB = int(input('Digite o 2° número: '))
if intA > intB:
    print(f'O maior número é {intA}')
elif intA < intB:
    print(f'O maior número é {intB}')
else:
    print(f'Os números são iguais!')
```
