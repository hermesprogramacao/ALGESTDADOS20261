```python
#1 - Peça ao usuário que digite sua idade usando input(), converta o valor para inteiro utilizando int() e exiba na tela a idade somada com 5 anos.
idade = int(input('Digite a sua idade: '))
print(f'{idade + 5} anos')
```

```python
#2 - Solicite ao usuário sua altura em metros, converta o valor digitado para float e mostre o dobro da altura informada.
altura = float(input('Digite sua altura (m): '))
print(f'O dobro da sua altura é: {(altura*2):.2f} m')
```

```python
#3 - Crie uma variável inteira com valor 100, converta esse valor para str e concatene com a frase " pontos alcançados", exibindo o resultado final.
valor = str(100)
print(f"{valor} pontos alcançados") #opção 1
print(valor + " pontos alcançados") #opção 2
```

```python
#4 - Peça ao usuário que digite 1 para verdadeiro ou 0 para falso, converta primeiro para inteiro e depois para bool, mostrando o valor booleano resultante.
opcao = bool(int(input("Digite 1 para verdadeiro ou 0 para falso: ")))
print(opcao)
```

```python
#5 - Peça três números ao usuário, converta todos para inteiro e mostre o maior entre eles.
num_A = int(input('Digite o primeiro número: '))
num_B = int(input('Digite o segundo número: '))
num_C = int(input('Digite o terceiro número: '))
print(f'Entre {num_A}, {num_B}, {num_C}, o maior número é {max(num_A,num_B,num_C)}')
```

```python
#6 - Peça quatro números decimais ao usuário, converta para float e calcule a média aritmética entre eles.
num_A = float(input('Digite o primeiro número: '))
num_B = float(input('Digite o segundo número: '))
num_C = float(input('Digite o terceiro número: '))
num_D = float(input('Digite o quarto número: '))
print(f'A média entre {num_A}, {num_B}, {num_C}, {num_D}, é {((num_A + num_B + num_C + num_D)/4):.2f}')
```

```python
# 7 - Peça a idade do usuário em anos, converta para inteiro e calcule aproximadamente quantos meses essa pessoa já viveu.
idade = int(input('Digite a sua idade: '))
idade = idade * 12
print(f'Essa pessoa já viveu {idade} meses, aproximadamente.')
```

```python
#8 - Peça ao usuário que digite um número inteiro, converta esse número para str e utilize a função len() para mostrar quantos caracteres foram digitados.
numero = str(int(input('Digite um número: ')))
print(f'O número {numero} tem {len(numero)} caracteres.')
```

```python
#9 - Escreva um pequeno código que tente somar 1 a um valor recebido por input() sem realizar conversão, observe o erro gerado e depois corrija o código realizando a conversão adequada.
numero = input('Digite um número: ') #String não pode ser somado com número inteiro
numero = int(numero) #Precisa fazer conversão para inteiro
numero += 1 #numero = numero + 1
print(numero)
```

```python
#10 - Peça um número decimal ao usuário, converta para float e depois converta esse valor para inteiro usando int(), exibindo ambos os resultados para comparar a diferença.
decimal = float(input('Digite um número decimal: '))
inteiro = int(decimal)
print(f'Decimal: {decimal} / Inteiro: {inteiro}')
```

```python
#11 - Solicite um número ao usuário, converta para inteiro e depois para bool, explicando em quais situações o resultado será False.
inteiro = int(input('Digite um número: '))
boolean = bool(inteiro)
print(f'{boolean}, pois apenas valores 0 ou vazio é False em Python')
```

```python
#12 - Peça ao usuário uma temperatura em Kelvin, converta para float, transforme o valor para Celsius e exiba o resultado.
kelvin = float(input('Digite uma temperatura em Kelvin: '))
print(f'A tempertura de {kelvin}K em Celsius é: {(kelvin-273.15):.2f}°C')
```

```python
#13 - Desenvolva um pequeno programa que receba nome, idade e salário do usuário, convertendo idade para inteiro e salário para float, e mostre todas as informações organizadas na tela.
nome = input('Digite seu nome: ')
idade = int(input('Digite sua idade: '))
salario = float(input('Digite seu salário: '))
print(f'--- FICHA CADASTRAL --- \nNome: {nome}\nIdade: {idade}\nSalário: R$ {salario}')
```

```python
#14 - Solicite o preço de um produto e o percentual de desconto, converta ambos para float, calcule o valor final com desconto aplicado e exiba o resultado.
preco = float(input('Digite o preço: R$'))
desconto = float(input('Digite o desconto(%): '))
preco = preco * ((100-desconto)/100)
print(f'O preço com desconto é: R${preco:.2f}')
```

```python
#15 - Peça ao usuário que digite um valor numérico, tente convertê-lo para inteiro e, caso ocorra erro de conversão, exiba uma mensagem informando que o valor digitado é inválido.
try:
  num = input('Digite um valor numérico: ') #Apenas números inteiros, caso seja decimal, tem que converter antes pra float
  convertido = int(num)
  print(convertido)
except:
  print('Valor inválido - Apenas números inteiros, caso seja decimal, tem que converter antes pra float')
```
