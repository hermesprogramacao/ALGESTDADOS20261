```python
#14 - Solicite o preço de um produto e o percentual de desconto, converta ambos para float, calcule o valor final com desconto aplicado e exiba o resultado.
produto = float(input('Digite o preço do produto: '))
desconto = float(input('Digite o desconto que será aplicado (%):'))
produto = produto * ((100 - desconto)/100)

print(f'O poreço do produto com desconto é {produto :.2f}')
```
