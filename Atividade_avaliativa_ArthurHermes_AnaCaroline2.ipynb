```python
#01 Levantamento Salarial
soma = 0
maior = 0
acima_3000 = 0

for i in range(1,11):
    salario = float(input(f'Digite o salário do {i}° funcionário: '))

    soma += salario

    if salario > maior:
        maior = salario

    if salario > 3000:
        acima_3000 += 1

media = soma / 10

print("Média:", media)
print("Maior salário:", maior)
print("Acima de 3000:", acima_3000)
```

```python
#02 Desempenho da Turma
soma = 0
# guarda a soma das notas

quantidade = 0
# conta quantos alunos foram digitados

aprovados = 0
# conta os alunos aprovados

reprovados = 0
# conta os alunos reprovados

nota = float(input("Digite a nota (-1 para parar): "))
# lê a primeira nota

while nota != -1:
# continua enquanto a nota for diferente de -1

    if nota < 0 or nota > 10:
        print("Digite uma nota entre 0 e 10.")
        # não aceita nota inválida

    else:
        soma = soma + nota
        # soma as notas

        quantidade = quantidade + 1
        # conta os alunos

        if nota >= 7:
            aprovados = aprovados + 1
            # conta aprovados
        else:
            reprovados = reprovados + 1
            # conta reprovados

    nota = float(input("Digite a nota (-1 para parar): "))
    # pede outra nota

if quantidade > 0:

    media = soma / quantidade
    # calcula a média

    print("Média:", media)
    print("Aprovados:", aprovados)
    print("Reprovados:", reprovados)

else:
    print("Nenhuma nota foi digitada.")
```

```python

#03 Sistema Bancário de Saques

saldo = float(input("Digite o saldo inicial: "))
# saldo da conta

saque = float(input("Digite o valor do saque (0 para sair): "))
# primeiro saque

while saque != 0:

    if saque < 0:
        print("Digite um valor válido.")
        # não aceita saque negativo

    elif saque <= saldo:
        saldo = saldo - saque
        # realiza o saque

        print("Saque realizado. Saldo atual:", saldo)

    else:
        print("Saldo insuficiente.")
        # não deixa sacar mais do que tem

    saque = float(input("Digite o valor do saque (0 para sair): "))
    # pede novo valor

print("Encerrado. Saldo final:", saldo)
```

```python
#04 Pesquisa Demográfica

soma = 0
# soma das idades

quantidade = 0
# quantidade de pessoas

maior = 0
# guarda a maior idade

maiores_18 = 0
# conta pessoas com mais de 18 anos

idade = int(input("Digite a idade (0 para parar): "))
# lê a primeira idade

while idade != 0:
# continua enquanto não digitar 0

    soma = soma + idade
    # soma as idades

    quantidade = quantidade + 1
    # conta as pessoas

    if idade > maior:
        maior = idade
        # atualiza a maior idade

    if idade > 18:
        maiores_18 = maiores_18 + 1
        # conta maiores de 18

    idade = int(input("Digite a idade (0 para parar): "))
    # pede outra idade

if quantidade > 0:

    media = soma / quantidade
    # calcula a média

    print("Média das idades:", media)
    print("Maior idade:", maior)
    print("Maiores de 18:", maiores_18)

else:
    print("Nenhuma idade foi digitada.")
```

```python
#05 Controle de Acesso

senha_correta = 2024
# guarda a senha correta do sistema

tentativas = 0
# contador de quantas tentativas já foram feitas

while tentativas < 3:
# o loop só roda enquanto o número de tentativas for menor que 3

    senha = int(input("Digite a senha: "))
    # pede a senha ao usuário

    if senha == senha_correta:
        # verifica se a senha digitada está correta

        print("Acesso liberado.")
        # mensagem de sucesso

        tentativas = 3
        # força a saída do loop (como se fosse um "parar")

    else:
        tentativas = tentativas + 1
        # soma 1 tentativa (porque errou)

        if tentativas == 3:
            # verifica se chegou no limite

            print("Acesso bloqueado.")
            # bloqueia após 3 erros

        else:
            print("Senha incorreta.")
            # avisa que errou, mas ainda tem chances
```

```python
#07 Análise de Sequência Numérica

pares = 0
# conta quantos números são pares

impares = 0
# conta quantos números são ímpares

soma_positivos = 0
# soma os números positivos

numero = int(input("Digite um número (0 para parar): "))
# lê o primeiro número

while numero != 0:
# continua enquanto o número for diferente de 0

    if numero % 2 == 0:
        pares = pares + 1
        # conta números pares
    else:
        impares = impares + 1
        # conta números ímpares

    if numero > 0:
        soma_positivos = soma_positivos + numero
        # soma apenas números positivos

    numero = int(input("Digite um número (0 para parar): "))
    # pede outro número

print("Quantidade de pares:", pares)
print("Quantidade de ímpares:", impares)
print("Soma dos positivos:", soma_positivos)
```

```python
#20 Estatísticas de Notas
soma = 0
# soma das notas

quantidade = 0
# conta quantas notas foram digitadas

abaixo_media = 0
# conta quantas notas ficaram abaixo da média

maior = 0
# guarda a maior nota

nota = float(input("Digite a nota (-1 para parar): "))
# lê a primeira nota

while nota != -1:
# continua enquanto a nota for diferente de -1

    soma = soma + nota
    # soma as notas

    quantidade = quantidade + 1
    # conta as notas

    if quantidade == 1:
        maior = nota
        # primeira nota vira a maior inicialmente

    elif nota > maior:
        maior = nota
        # atualiza a maior nota

    nota = float(input("Digite a nota (-1 para parar): "))
    # pede outra nota

if quantidade > 0:

    media = soma / quantidade
    # calcula a média

    nota = float(input("Digite novamente as notas para verificar abaixo da média (-1 para parar): "))
    # lê novamente as notas

    while nota != -1:

        if nota < media:
            abaixo_media = abaixo_media + 1
            # conta notas abaixo da média

        nota = float(input("Digite novamente as notas para verificar abaixo da média (-1 para parar): "))

    print("Média:", media)
    print("Maior nota:", maior)
    print("Notas abaixo da média:", abaixo_media)

else:
    print("Nenhuma nota foi digitada.")
```

```python
#27 Par ou Ímpar até N

n = int(input("Digite um número inteiro: "))
# lê o valor de N

for i in range(1, n + 1):
# percorre os números de 1 até N

    if i % 2 == 0:
        # verifica se o número é par

        print(i, "- Par")

    else:
        # se não for par, é ímpar

        print(i, "- Ímpar")
```

```python
#28 Caça aos Primos

quantidade_primos = 0
# conta quantos números são primos

soma_nao_primos = 0
# soma os números que não são primos

numero = int(input("Digite um número (-1 para parar): "))
# lê o primeiro número

while numero != -1:
# continua enquanto o número for diferente de -1

    divisores = 0
    # conta quantos divisores o número possui

    for i in range(1, numero + 1):
    # verifica todos os divisores possíveis

        if numero % i == 0:
            divisores = divisores + 1
            # conta divisores exatos

    if divisores == 2:
        quantidade_primos = quantidade_primos + 1
        # número primo possui exatamente 2 divisores

    else:
        soma_nao_primos = soma_nao_primos + numero
        # soma números que não são primos

    numero = int(input("Digite um número (-1 para parar): "))
    # pede outro número

print("Quantidade de números primos:", quantidade_primos)
print("Soma dos números não primos:", soma_nao_primos)
```

```python
#29 Avaliação de Notas

soma = 0
# soma das notas

quantidade = 0
# conta quantas notas foram digitadas

acima_media = 0
# conta notas acima da média

nota = float(input("Digite a nota (0 para parar): "))
# lê a primeira nota

while nota != 0:
# continua enquanto a nota for diferente de 0

    soma = soma + nota
    # soma as notas

    quantidade = quantidade + 1
    # conta as notas

    nota = float(input("Digite a nota (0 para parar): "))
    # pede outra nota

if quantidade > 0:

    media = soma / quantidade
    # calcula a média

    nota = float(input("Digite novamente as notas (0 para parar): "))
    # lê novamente as notas

    while nota != 0:

        if nota > media:
            acima_media = acima_media + 1
            # conta notas acima da média

        nota = float(input("Digite novamente as notas (0 para parar): "))

    print("Média geral:", media)
    print("Notas acima da média:", acima_media)

else:
    print("Nenhuma nota foi digitada.")
```

```python
#30 Cálculo de Fatorial

numero = int(input("Digite um número inteiro positivo: "))
# lê o número

fatorial = 1
# começa o fatorial em 1

for i in range(1, numero + 1):
# percorre os números de 1 até o número digitado

    fatorial = fatorial * i
    # multiplica os valores para calcular o fatorial

print("Fatorial:", fatorial)

if fatorial % 2 == 0:
    # verifica se o fatorial é par

    print("O fatorial é par.")

else:
    # se não for par, é ímpar

    print("O fatorial é ímpar.")
```
